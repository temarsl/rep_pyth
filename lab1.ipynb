```python
import math

def task1(x):
    d = x * x - 8 * x + 12
    if d == 0:
        return None
    return (x * x - 7 * x + 10) / d

task1(3)
```

```python
def task2(x):
    d = 1 - math.cos(2 * x) + math.sin(2 * x)
    if d == 0:
        return None
    return (1 + math.cos(2 * x) + math.sin(2 * x)) / d

task2(1)
```

```python
def distance(x1, y1, x2, y2): # 
    return math.hypot(x2 - x1, y2 - y1)

def triangle_perimeter(x1, y1, x2, y2, x3, y3): # 
    a = distance(x1, y1, x2, y2)
    b = distance(x2, y2, x3, y3)
    c = distance(x3, y3, x1, y1)
    return a + b + c

def triangle_area(x1, y1, x2, y2, x3, y3): #
    area = 0.5 * abs(
        x1*(y2 - y3) +
        x2*(y3 - y1) +
        x3*(y1 - y2)
    )
    return area

def task3(x1, y1, x2, y2, x3, y3):
    a1 = (triangle_perimeter(x1, y1, x2, y2, x3, y3))
    a2 = (triangle_area(x1, y1, x2, y2, x3, y3))
    return a1, a2

task3(1,2,5,8,9,20)
```

```python
def can_move(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    return (dx == 2 and dy == 1) or (dx == 1 and dy == 2)

def task5(x1,x2,y1,y2):
    if can_move(x1, y1, x2, y2):
        return ("     .")
    else:
        return ("      .")

task5(4,8,10,20)
```

```python
import math

def area_base_height(base, height):
    return 0.5 * base * height

def area_two_sides_angle(a, b, angle_rad):
    return 0.5 * a * b * math.sin(angle_rad)

def area_three_sides(a, b, c):
    s = (a + b + c) / 2
    #  
    return math.sqrt(s * (s - a) * (s - b) * (s - c))

def task1():
    while True:
        print("\n   :")
        print("1 -   ")
        print("2 -       ( )")
        print("3 -  ")
        print("0 - ")

        group = int(input("  : "))

        if group == 0:
            print(" .")
            break
        elif group == 1:
            base = float(input(" : "))
            height = float(input(" : "))
            area = area_base_height(base, height)
            print(f" : {area:.2f}")
        elif group == 2:
            a = float(input("  : "))
            b = float(input("  : "))
            angle = float(input("    ( ): "))
            area = area_two_sides_angle(a, b, angle)
            print(f" : {area:.2f}")
        elif group == 3:
            a = float(input("  : "))
            b = float(input("  : "))
            c = float(input("  : "))
            # ,   
            if a + b > c and a + c > b and b + c > a:
                area = area_three_sides(a, b, c)
                print(f" : {area:.2f}")
            else:
                print("     .")
        else:
            print("  .  .")

task1()
```

```python
def calculate_s(n):
    s = 0
    for i in range(n + 1):
        term = (-1) ** i * (1 / (2 ** i))
        s += term
    return s

def task2(n):
    if n <= 0:
        print(",    (n > 0).")
    else:
        result = calculate_s(n)
        print(f" S  n = {n} : {result}")

task2(1)
```

```python
def calculate_series_epsilon(epsilon):
    s = 0
    n = 1
    term = 1.0  #   
    while abs(term) >= epsilon:
        s += term
        n += 1
        term = (-1) ** (n - 1) / n
    return s, n - 1

def task3():
    #  
    epsilon = 0.001

    #  
    sum_result, terms = calculate_series_epsilon(epsilon)
    print(f"    {epsilon} = {sum_result}")
    print(f" : {terms}")

task3()
```

```python
def create_digit_array(n):
    return [int(digit) for digit in str(n)]

def task1():
    n = int(input("   N: "))

    # ,   
    if n <= 0:
        print("   (N > 0)")
    else:
        #     
        digits = create_digit_array(n)
        print(f"   {n}: {digits}")

task1()
```
